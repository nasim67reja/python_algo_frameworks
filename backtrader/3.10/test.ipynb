```python
import matplotlib.pyplot as plt


# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Create a plot
plt.plot(x, y)

# Add title and labels
plt.title("Simple Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Show the plot
plt.show()
```
